```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Names = ['All_attributes','equipment', 'water', 'grazing', 'milk', 'contact_animal', 'contact_human', 'vet']
```

```python
def summarystats(name):
    param = ['Total Number of Infections', 'Endemic Level']
    df = pd.DataFrame(columns = ['Farms', 'Summary'])
    ENumber = np.load('Endemic_{}.npy'.format(name))
    CNumber = np.load('Size_{}.npy'.format(name))
    for i in range(len(param)):
        df3 = pd.DataFrame()
        if i ==0:
            df3['Farms'] = CNumber
            df3['Summary'] = param[i]
        if i ==1:
            df3['Farms'] = ENumber
            df3['Summary'] = param[i]
        df = df.append(df3, ignore_index = True)
    df['Model'] = name
        
    return df
```

```python
D0 = summarystats(Names[0])
D1 = summarystats(Names[1])
D2 = summarystats(Names[2])
D3 = summarystats(Names[3])
D4 = summarystats(Names[4])
D5 = summarystats(Names[5])
D6 = summarystats(Names[6])
D7 = summarystats(Names[7])
```

```python
# D1
```

```python
Df = D0.append([D0, D1, D2, D3, D4, D5, D6, D7])
```

```python
plt.figure(figsize = (10, 4))
sns.boxplot(x = 'Summary', y = 'Farms', data = Df.loc[Df['Summary'].isin(['Total Number of Infections'])], hue = 'Model' )
# plt.yscale('log')
plt.legend(loc = 'lower left', ncol =2)
plt.ylim([9000, 20000])
# plt.ylabel('Percentage %')
```

```python
plt.figure(figsize = (10, 4))
sns.boxplot(x = 'Summary', y = 'Farms', data = Df.loc[Df['Summary'].isin(['Endemic Level'])], hue = 'Model' )
plt.ylim([100, 250])
plt.legend(loc = 'lower left', ncol = 2)
```
